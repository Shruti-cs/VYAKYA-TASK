```python
#IMPORT LIBRARIES:
import json
import pandas as pd
```

```python
# OPEN JSON FILE:
json_data = open("C:/Users/hp/Desktop/Technical-Assessment/task2/task_input_list.json")
  
# RETURN JSON OBJECT AS A DICTIONARY:
data = json.load(json_data)
```

```python
#DISPLAY DATA:
data2=data['data']
for i in range(len(data2)):
    print(data2[i])
```

```python
#PERFORMING VALIDATIONS ON DATE AND AMOUNT COLUMNS:
amt=[]
for i in range(len(data2)):
    try:
        if data2[i][2]=='/':
            listt=[]
            listt.append(data2[i])
            listt.append(data2[i+1])
            listt.append(data2[i+2])
            amt.append(listt)
    except:
        print('*******')
print(amt)
```

```python
#IDENTOFYING PHONE NUMBERS AND EMAIL ID IN THE LIST:
#PHONE NUMBER:
phn=[]
for i in data2:
    if i.count('-')==3 and i.count(':')==1:
        a=i
        ph=''
        for j in a:
            if j.isnumeric() or (j=='-' or j=='(' or j==')'):
                ph=ph+j
        phn.append(ph)
    
    if i.count('-')>=1 and i.count('(')==1:
        a=i
        ph=''
        for j in a:
            if j.isnumeric() or (j=='-' or j=='(' or j==')'):
                ph=ph+j
        phn.append(ph)
print(phn)
```

```python
#EMAIL ID:
email_id=""
for i in data2:
    a=i
    if "@" in a:
        if ".com" in a:
            email_id=email_id+i
            break
    else:
        email_id="NA"
email_id
```

```python
#WEBSITE:
website=""
for i in data2:
    a=i
    if ".com" in a:
        website=website+i
        break
website        
```

```python
#IDENTIFYING DEPOSIT TRANSACTION LIST AND WITHDRAWAL TRANSACTION LIST:
d_list=[]
w_list=[]
for i in amt:
    amount=i[2]
    da=i[0]
    if amount[0]!='-':
        deposit=[]
        deposit.append(i[0])
        deposit.append(i[1])
        deposit.append(amount)
        deposit.append(da[0:2])
        deposit.append(da[3:5])
        deposit.append(da[6::])
        d_list.append(deposit)
    else:
        withdraw=[]
        withdraw.append(i[0])
        withdraw.append(i[1])
        withdraw.append(amount)
        withdraw.append(da[0:2])
        withdraw.append(da[3:5])
        withdraw.append(da[6::])
        w_list.append(withdraw) 
print(d_list)
print(w_list)
```

```python
deposit_df=pd.DataFrame(d_list,columns=['Date','Description','Amount','month','day','year'])
deposit_df
```

```python
deposit_df.info()
```

```python
withdraw_df=pd.DataFrame(w_list,columns=['Date','Description','Amount','month','month','year'])
withdraw_df
```

```python
withdraw_df.info()
```

```python
#Converting Amount column form string to Float

withdraw_df['Amount'] = withdraw_df['Amount'].str.replace(',', '').astype(float)
```

```python
#FINDING MAXIMUN AMOUNT AND MINIMUM AMOUNT:
#MAX AMOUNT:
col = deposit_df["Amount"]
max_amt = col.max()
max_amt
```

```python
#MIN AMOUNT:
col = withdraw_df["Amount"]
min_amt = col.min()
min_amt
```
